```python
import pandas as pd
import glob
```

```python
path=r'C:\Users\dell\Documents\305 Crestview\305 Analysis\Data\Raw Abb Tx Hx Dnlds'
all_files=glob.glob(path + "/airbnb_t*") + glob.glob(path + "/airbnb_tax*")

li=[]

for filename in all_files:
    df=pd.read_csv(filename
                   ,header=0
                   ,encoding='utf8')
    df.columns = df.columns.str.title()
    li.append(df)
    
frame=pd.concat(li,axis=0,ignore_index=True,sort=False)
```
